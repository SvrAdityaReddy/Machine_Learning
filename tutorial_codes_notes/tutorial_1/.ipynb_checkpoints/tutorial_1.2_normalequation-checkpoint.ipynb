```python
#http://www.ozzieliu.com/tutorials/Linear-Regression-Gradient-Descent.html
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('../tutorial/datasets/housing/housing_subset.csv')
data.head()
```

```python
#X_df = data['median_income']
X_df = pd.DataFrame(data.median_income)
y_df = pd.DataFrame(data.median_house_value)
#y_df = data['median_house_value']
m = len(y_df)
```

```python
plt.figure(figsize=(10,8))
plt.plot(np.array(X_df), np.array(y_df), 'kx')
plt.xlabel('median_income')
plt.ylabel('median_house_value')

```

```python
plt.figure(figsize=(10,8))
plt.plot(X_df, y_df, 'k.')
plt.plot([-2, 15], [120000,150000], '-')
plt.plot([-2, 14], [-1000,500000], '-')
plt.plot([-2, 8], [-80000,600000], '-')
plt.xlabel('median_income')
plt.ylabel('median_house_value')
```

```python
iterations = 1500
alpha = 0.01
```

```python
## Add a columns of 1s as intercept to X
X_df['intercept'] = 1

## Transform to Numpy arrays for easier matrix math and start theta at 0
X = np.array(X_df)
y = np.array(y_df).flatten()
theta = np.array([0, 0])
```

```python
def cost_function(X, y, theta):
    """
    cost_function(X, y, theta) computes the cost of using theta as the
    parameter for linear regression to fit the data points in X and y
    """
    ## number of training examples
    m = len(y) 
    
    ## Calculate the cost with the given parameters
    J = np.sum((X.dot(theta)-y)**2)/m
    
    return J
```

```python
cost_function(X, y, theta)
```

```python
def gradient_descent(X, y, theta, alpha, iterations):
    
    cost_history = [0] * iterations
    iter_num = [0] * iterations
    theta_history = [0] * iterations
    
    for iteration in range(iterations):
        hypothesis = X.dot(theta)
        error = hypothesis-y
        gradient = 2/m* X.T.dot(error)
        theta = theta - alpha*gradient
        theta_history[iteration] = theta
        cost = cost_function(X, y, theta)
        cost_history[iteration] = cost
        iter_num[iteration] = iteration
        
    return theta, cost_history, iter_num, theta_history
     
```

```python
(t, c, num, t_his) = gradient_descent(X,y,theta,alpha, iterations)
t
```

```python
plt.figure(figsize=(5,4))
plt.plot(num[0:100], c[0:100], '-')

plt.xlabel('iteration')
plt.ylabel('Cost')
```

```python
## Plotting the best fit line

best_x = np.linspace(0, 25, 20)
best_y1 = [t_his[0][1] + t_his[0][0]*xx for xx in best_x]
best_y0 = [t_his[1][1] + t_his[1][0]*xx for xx in best_x]
best_y2 = [t_his[2][1] + t_his[2][0]*xx for xx in best_x]
best_y3 = [t_his[4][1] + t_his[4][0]*xx for xx in best_x]
best_y31 = [t_his[5][1] + t_his[5][0]*xx for xx in best_x]
best_y4 = [t_his[6][1] + t_his[6][0]*xx for xx in best_x]
best_y5 = [t_his[8][1] + t_his[8][0]*xx for xx in best_x]
best_y6 = [t_his[10][1] + t_his[10][0]*xx for xx in best_x]
best_y7 = [t_his[15][1] + t_his[15][0]*xx for xx in best_x]
best_y8 = [t_his[20][1] + t_his[20][0]*xx for xx in best_x]
best_y9 = [t_his[100][1] + t_his[100][0]*xx for xx in best_x]
```

```python
plt.figure(figsize=(10,6))
plt.plot(X_df.median_income, y_df, '.')
plt.plot(best_x, best_y1, '-')
plt.plot(best_x, best_y0, '-')
plt.plot(best_x, best_y2, '-')
plt.plot(best_x, best_y3, '-')
plt.plot(best_x, best_y31, '-')
plt.plot(best_x, best_y4, '-')
plt.plot(best_x, best_y5, '-')
plt.plot(best_x, best_y6, '-')
plt.plot(best_x, best_y7, '-')
plt.plot(best_x, best_y8, '-')
plt.plot(best_x, best_y9, '*')
plt.axis([-2,16,0,550000])
plt.xlabel('median_income')
plt.ylabel('median_house_value')
plt.title('Linear Regression Line')
```

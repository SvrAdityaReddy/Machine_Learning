```python
#https://github.com/harihari1989/LinearRegression
import pandas as pd
from numpy.linalg import inv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('../tutorial/datasets/housing/housing_subset.csv')
data.head()
```

```python
#X_df = data['median_income']
X_df = pd.DataFrame(data.median_income)
y_df = pd.DataFrame(data.median_house_value)
#y_df = data['median_house_value']
m = len(y_df)
```

```python
plt.figure(figsize=(10,8))
plt.plot(X_df, y_df, 'k.')
plt.plot([-2, 15], [120000,150000], '-')
plt.plot([-2, 14], [-1000,500000], '-')
plt.plot([-2, 8], [-80000,600000], '-')
plt.xlabel('median_income')
plt.ylabel('median_house_value')
```

```python

def normal_equation(x,y):
    z1 = np.dot(x.transpose(), x) 
    z = np.linalg.inv(z1)
    z2 = np.dot(z, x.transpose())
    theta = np.dot(z2, y)
    return theta
```

```python
X_df['intercept'] = 1
X = np.array(X_df)
y = np.array(y_df).flatten()
theta = np.array([0, 0])
```

```python
theta = normal_equation(X,y)
```

```python
theta
```

```python
## Prediction
print(np.array([4.9375, 1]).dot(theta))
print(np.array([3.3841, 1]).dot(theta))
#227700 184400
```

```python
## Plotting the best fit line
best_fit_x = np.linspace(0, 25, 20)
best_fit_y = [theta[1] + theta[0]*xx for xx in best_fit_x]
print(best_fit_x)
print(best_fit_y)
```

```python
plt.figure(figsize=(10,6))
plt.plot(X_df.median_income, y_df, '.')
plt.plot(best_fit_x, best_fit_y, '-')
plt.axis([-2,16,0,550000])
plt.xlabel('median_income')
plt.ylabel('median_house_value')
plt.title('Linear Regression Line')
```
